```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

result = str1 + " " + str2

print("\nConcatenated String :", result)

print("lower()      :", result.lower())
print("upper()      :", result.upper())
print("title()      :", result.title())
print("swapcase()   :", result.swapcase())
print("capitalize() :", result.capitalize())
print("casefold()   :", result.casefold())
print("center()     :", result.center(50, '*'))
print("count('a')   :", result.count('a'))
print("endswith()   :", result.endswith('a'))
print("find('a')    :", result.find('a'))
print("isalnum()    :", result.isalnum())
print("isdigit()    :", result.isdigit())
print("isnumeric()  :", result.isnumeric())
print("isspace()    :", result.isspace())
print("replace()    :", result.replace('a', '@'))
```

```python
a = 10
print("Initial value of a =", a)

a += 5
print("After += 5 :", a)

a -= 3
print("After -= 3 :", a)

a *= 2
print("After *= 2 :", a)

a /= 4
print("After /= 4 :", a)

a %= 3
print("After %= 3 :", a)

a **= 2
print("After **= 2 :", a)

a //= 2
print("After //= 2 :", a)
```

```python
name = input("Enter student name: ")
student_class = input("Enter class: ")

m1 = float(input("Enter marks of Subject 1: "))
m2 = float(input("Enter marks of Subject 2: "))
m3 = float(input("Enter marks of Subject 3: "))
m4 = float(input("Enter marks of Subject 4: "))
m5 = float(input("Enter marks of Subject 5: "))

total = m1 + m2 + m3 + m4 + m5
percentage = total / 5
if percentage >= 60:
    grade = 'A'
elif percentage >= 50:
    grade = 'B'
elif percentage >= 40:
    grade = 'C'
elif percentage >= 33:
    grade = 'D'
else:
    grade = 'Fail'

print("\n----- RESULT -----")
print("Name :", name)
print("Class :", student_class)
print("Total Marks :", total)
print("Percentage :", percentage, "%")
print("Grade :", grade)
```
