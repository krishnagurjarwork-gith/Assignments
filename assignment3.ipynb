```python
def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(maximum(10, 25, 15))
```

```python
def distinct_list(lst):
    new_list = []

    for item in lst:
        if item not in new_list:
            new_list.append(item)

    return new_list

print(distinct_list([1, 2, 2, 3, 4, 4, 5]))
```

```python
def multiply_list(lst):
    result = 1

    for num in lst:
        result = result * num

    return result

print(multiply_list([2, 3, 4]))
```

```python
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact = fact * i

    return fact

print(factorial(5))
```

```python
def reverse_string(s):
    return s[::-1]

print(reverse_string("Python"))
```

```python
def check_range(num, start, end):
    if num in range(start, end + 1):
        print("Number is within the range")
    else:
        print("Number is outside the range")

check_range(5, 1, 10)
```

```python
def even_numbers(lst):
    for num in lst:
        if num % 2 == 0:
            print(num)

even_numbers([1, 2, 3, 4, 5, 6, 7, 8])
```

```python
def prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

print(prime(7))
```

```python
def count_case(s):
    upper = 0
    lower = 0

    for ch in s:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

count_case("Hello Python")
```
