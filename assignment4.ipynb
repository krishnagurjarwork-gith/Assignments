```python
import csv

with open("address_book.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Address", "Mobile", "Email"])

    for i in range(3):
        print(f"\nEnter details of person {i+1}")

        name = input("Enter Name: ")
        address = input("Enter Address: ")
        mobile = input("Enter Mobile: ")
        email = input("Enter Email: ")

        writer.writerow([name, address, mobile, email])

print("\nData stored successfully in address_book.csv")
```

```python
import sqlite3

conn = sqlite3.connect("assignment.db")

cur = conn.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS Student(
    id INTEGER PRIMARY KEY,
    name TEXT,
    course TEXT
)
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS Teacher(
    id INTEGER PRIMARY KEY,
    name TEXT,
    subject TEXT
)
""")

cur.execute("INSERT INTO Student VALUES(1, 'Krishna', 'BCA')")
cur.execute("INSERT INTO Student VALUES(2, 'Rahul', 'MCA')")

cur.execute("INSERT INTO Teacher VALUES(1, 'Amit Sir', 'Python')")
cur.execute("INSERT INTO Teacher VALUES(2, 'Neha Maam', 'DBMS')")

conn.commit()

print("\nStudent Table:")
cur.execute("SELECT * FROM Student")

for row in cur.fetchall():
    print(row)

cur.execute("UPDATE Student SET course='B.Tech' WHERE id=1")
conn.commit()

print("\nAfter Update:")
cur.execute("SELECT * FROM Student")

for row in cur.fetchall():
    print(row)

cur.execute("DELETE FROM Student WHERE id=2")
conn.commit()

print("\nAfter Delete:")
cur.execute("SELECT * FROM Student")

for row in cur.fetchall():
    print(row)

conn.close()
```
